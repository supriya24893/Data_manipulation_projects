### 6. Create a new column Qtr using numpy.where() or any suitable utility in the imported dataset. The Quarters are based on months and defined as -
### •	Jan - Mar : Q1

### •	Apr - Jun : Q2

### •	Jul - Sep : Q3

### •	Oct - Dec : Q4   


```python
def quarter(x):
    if "Jan" in x or "Feb" in x or "Mar" in x:
        return 'Q1'
    elif "Apr" in x or "May" in x or "Jun" in x:
        return 'Q2'
    elif "Jul" in x or "Aug" in x or "Sep" in x:
        return 'Q3'
    elif "Oct" in x or "Nov" in x or "Dec" in x:
        return 'Q4'


sales["Qtr"] = sales["Month"].apply(quarter)
sales
```

### 7. Compare Qtr wise sales in 2015 and 2016 in a bar plot


```python
qtr_wise=sales.groupby(by=["Qtr"])
N=qtr_wise[["Sales2015","Sales2016"]].sum().reset_index()
N=pd.melt(N, id_vars="Qtr", var_name="Total_sales", value_name="Sales_by_year")
N
sns.factorplot(x='Qtr', y='Sales_by_year', hue='Total_sales', data=N, kind='bar')
```

### 8. Determine the composition of Qtr wise sales in and 2016 with regards to all the Tiers in a pie chart.
