### Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

```python
data=pd.read_csv("C:/Users/user/Desktop/Gurgaon course/Python_homeworks/Pandas Exercises/Pandas Exercises/Exercise Data Files/Pandas Exercise Date Time Operations Files/wind.data",sep=",")

#converting the 3 columns to string dtype
data.Yr.astype(str)
data.Mo.astype(str)
data.Dy.astype(str)

#concatenating the 3 columns into one column
new_date=data.Yr.astype(str)+'-'+data.Mo.astype(str)+'-'+data.Dy.astype(str)

#new_date is the concatenated column
data["New_date"]=new_date

#dropping the old 3 date columns
data=data.drop(columns=["Yr","Mo","Dy"])
data
```

### Step 4. Year 2061 is seemingly imporoper. Convert every year which are < 70 to 19XX instead of 20XX.

```python
#converting the column to datetime dtype
data["New_date"]=pd.to_datetime(data.New_date,format="%y-%m-%d")
data

#subtracting 100 years from all dates greater than 2018 by using the timedelta function
data['New_date'] = data['New_date'].mask(data['New_date'].dt.year > 2018, data['New_date'] - pd.to_timedelta(100, unit='y'))
data
```

### Step 16. Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.

```python
#resample it weekly and calc all parameters
u = data.New_date.min()+pd.DateOffset(day=1)
end=u+pd.DateOffset(weeks=52)
end
#starting from january 2nd and ending 52 weeks later
mask = (data['New_date']>= u) & (data['New_date'] <= end)
data=data.loc[mask]
data
#setting date as new index
data2=data.set_index("New_date")

#resampling it weekly and calculating mean,min,max and standard deviation
data2.resample("W").mean()
data2.resample("W").min()
data2.resample("W").max()
data2.resample("W").std()
```
