## Step 5. Transform the Date column as a datetime type

```python
#converting date column in apple dataframe  to datetime format
apple["Date"]=pd.to_datetime(apple["Date"],format="%Y-%m-%d")
apple["Date"]
```

### Step 6.  Set the date as the index

```python

new_index=apple.set_index("Date")
```

### Step 7.  Is there any duplicate dates?

```python
new_index.duplicated().value_counts()
```

### Step 8.  The index is from the most recent date. Sort the data so that the first entry is the oldest date.

```python
new_index.sort_values(by=["Date"],ascending=True)
```

### Step 9. Get the last business day of each month

```python
#finds the earliest date from column date
start = apple.Date.min()

#finds the latest date from column date
end = apple.Date.max()

#gives the daterange between the start and end end. BM provides frequency which is business month and calculates last business day of each month
business_days_rng =pd.date_range(start, end, freq='BM')
business_days_rng
```

### Step 10.  What is the difference in days between the first day and the oldest

```python
#calculates the max and min date and subtract the difference
difference=apple.Date.max()-apple.Date.min()
difference
```

## Step 11.  How many months in the data we have?

```python
apple["Months"]=apple.Date.dt.month
apple
apple.Months.nunique()
```
