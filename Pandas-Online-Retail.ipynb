### Step 5. Create a scatterplot with the Quantity per UnitPrice by CustomerID for the top 3 Countries

```python
pd.to_numeric(online_rt.Quantity)
pd.to_numeric(online_rt.UnitPrice)
online_rt["Quantityperunitprice"]=online_rt.Quantity/online_rt.UnitPrice
online_rt

```

```python
a=online_rt.sort_values(by=['Quantity','Country'],ascending=False)
a[a.replace([np.inf, -np.inf], np.nan).notnull().all(axis=1)]
a=a.drop_duplicates("Country")

b=a.sort_values(['Quantityperunitprice'], ascending=False).head(3)
b
sns.lmplot(x="CustomerID",y="Quantityperunitprice",data=b,fit_reg=False,hue="Country")
plt.show()
```
