### Step 3. Assign it to a variable called users.

```python
#assigning it to variable users
users=pd.read_csv("C:/Users/user/Desktop/Gurgaon course/Python_homeworks/Pandas Exercises/Pandas Exercises/Exercise Data Files/Pandas Exercise Aggregation Files/occupation.csv")

#splitting the one column into multiple columns
users['index'], users['age'],users['gender'], users['occupation'],users['zipcode'] = users['user_id|age|gender|occupation|zip_code'].str.split('|', 4).str
users=users.drop(columns=["user_id|age|gender|occupation|zip_code"])

#Making the dtype of the dataframe users columns age and gender as integer and string respectively
users["age"]=users.age.astype(int)
users["gender"]=users.gender.astype(str)
users
```

### Step 4. Discover what is the mean age per occupation

```python
#grouping by occupation
temp=users.groupby(by=["occupation"])

#finding mean age per group
temp[["age"]].mean().add_prefix("Mean ").reset_index()
```

## Step 5.Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
#encoding the gender column with males=0,females=1
users["encoded_gender"]=np.where((users["gender"]=="M"),0,np.where((users["gender"]=="F"),1," "))
users

#finding total number of people in each occupation
occupation_wise_people=users.groupby(by=["occupation"])[["encoded_gender"]].count().add_prefix("total ").reset_index()

#filtering for male gender
only_males=users.loc[(users.encoded_gender=='0')]
#grouping only_males by occupation and counting the number of males
only_males1=only_males.groupby(by=["occupation"])[["encoded_gender"]].count().add_prefix("number of males ").reset_index()

#merging 2 tables
males_ratio=pd.merge(occupation_wise_people,only_males1,on='occupation',how='left')

#dividing the two columns to get the male ratio
males_ratio["M_Ratio"]=males_ratio["number of males encoded_gender"]/males_ratio["total encoded_gender"]
males_ratio["males_percent"]=males_ratio["M_Ratio"]*100
males_ratio
```

## Step 6. For each occupation, calculate the minimum and maximum ages

```python
#grouping by occupation
temp=users.groupby(by=["occupation"])

#calculating the minimum and maximum age 
res1=temp[["age"]].max().add_prefix("Max_age ").reset_index()
res2=temp[["age"]].min().add_prefix("Min_age ").reset_index()

#concatenating columns
pd.concat([res1,res2],axis=1)
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
#grouping by occupation and gender
temp3=users.groupby(by=["occupation","gender"])

#calculating mean age per occupation per gender
temp3[["age"]].mean().add_prefix("Mean_of_").reset_index()
```

### Step 8.  For each occupation present the percentage of women and men

```python
#filtering for female gender
only_females=users.loc[(users.encoded_gender=='1')]

#counting number of females
only_females1=only_females.groupby(by=["occupation"])[["encoded_gender"]].count().add_prefix("number of females ").reset_index()

#merging both tables
females_ratio=pd.merge(left=males_ratio,right=only_females1,on='occupation',how='left')
females_ratio

#finding percent ratio
females_ratio["F_percent_ratio"]=females_ratio["number of females encoded_gender"]*100/females_ratio["total encoded_gender"]
females_ratio
```
